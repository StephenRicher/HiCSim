```python
import numpy as np
import matplotlib
import random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
matplotlib.rcParams['figure.figsize'] = [16, 14]
```

```python
n_beads  = 4400
bases_per_bead = 1000

# Randomly select rosette length between 30kbp - 100kbp
rosette_length = random.randint(30,100) * 1000
beads_per_loop = rosette_length / bases_per_bead
# Randomly select number of loops per turn between 4 and 12
loops_per_turn = random.randint(4,12)

beads_per_turn = beads_per_loop * loops_per_turn
n_turns = n_beads / beads_per_turn

k = loops_per_turn / 2
vx = 0.38
print(k, n_turns)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

theta = np.linspace(0, n_turns * 2 * np.pi, n_beads)

x = 12 * (vx + (1 - vx) * (np.cos(k * theta))**2 * np.cos(theta))
y = 12 * (vx + (1 - vx) * (np.cos(k * theta))**2 * np.sin(theta))
z = theta * 2 / (2 * np.pi)

ax.plot(x, y, z)
plt.show()
```
